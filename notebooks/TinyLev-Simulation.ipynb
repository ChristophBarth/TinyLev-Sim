# Simulation des akustischen Levitators "Tiny-Lev"
Ein akustischer Levitator besteht aus einem oder mehreren piezoelektrischen Elementen, im Folgenden Piezo genannt. Diese haben eine oszillierende Membran, die Schallwellen erzeugt. Ein akustischer Levitator nutzt die Felder dieser Schallwellen aus, um Objekte schweben zu lassen. Feldeigenschaft ist hierbei der Luftdruck $P$, der durch die Membran verändert wird.
Das akustische Feld im Punkt $p$, verursacht durch eine Punktquelle, lässt sich mit der Wellenfunktion $\Psi$ beschreiben
$\Psi(p)=sin(t+K*d)$
wobei $t$ die gegenwärtige Phase des Piezos beschreibt und $d$ der Abstand des Piezos zum Punkt$p$. $K=\frac{2\pi}{\lambda}$ die Wellenzahl ist.

Die Näherung des Piezos als Punktquelle ist physikalisch allerdings unakkurat, wehalb diese Wellenfunktion erweitert werden muss:
1. Die Stärke des akustischen Feldes verhält sich antiproportional zur Distanz vom Piezo zum Punkt $p$. Die Gleichung muss somit mit einem Faktor $\frac1d$ erweitert werden.
2. Der Piezo erzeugt ein gerichtetes Druckfeld. Die Senderichtung eines Piezos wird durch einen Normalenvektor $\vec{n}$ beschrieben, der orthogonal aus dessen Membran steht. Je größer der Winkel zwischen der Normalen und dem Punkt ist, desto kleiner wird das Feld. So muss eine Direktivitätsfunktion $D_f(\theta)$ definiert, die vom Winkel $\theta$ zwischen Normalen und Punkt abhängt.

## Direktivitätsfunktion $D_f(\theta)$
Der Winkel $\theta$ lässt sich über das Skalarprodukt der Piezonormalen $\vec{n}$ und des Vektors $\vec{Tp}$ vom Piezo zum Punkt $p$ berechnen. Es gilt
$\theta = arccos(\frac{\vec{n} * \vec{Tp}}{|\vec{n}| * |\vec{Tp}|})$
Die Direktivitätsfunktion für einen Piezo lautet
$D_f(\theta) = \frac{2*J_1(k*a*sin(\theta)}{k*a*sin(\theta)}$
mit $J_1$ als erste Bessel Funktion erster Ordnung und dem Membranradius $a$.
